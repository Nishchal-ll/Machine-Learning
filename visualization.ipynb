## Why Choose matplotlib and seaborn?

Both libraries are among the most popular choices for data visualization in Python. Hereâ€™s why:

Matplotlib: Known for its flexibility, Matplotlib allows detailed customization of visual elements, making it perfect for tailored plots.

Seaborn: Built on top of Matplotlib, Seaborn simplifies visualization tasks by offering high-level functions to create aesthetically pleasing statistical graphics.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# basics of marplotlib
#line plot

x=np.linspace(0,500,5)
y=np.sin(x)
plt.plot(x,y)
```

```python
x=np.linspace(0,500,5)
y=np.sin(x)
plt.plot(x,y, label="Sine-wave")
plt.title("Line plot example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# by chart

categories=['A','B','C']
values = [20,10,20]
plt.bar(categories,values,color=['white','pink','white'])
plt.title("Bar chart example")
plt.xlabel("Categories")
plt.ylabel("Values")
ax = plt.gca()
ax.set_facecolor('black')
plt.show()
```

```python
import seaborn as sns
```

```python
# histogram and KDE plot

data=np.random.randn(1000)
ax= sns.histplot(data,kde=True,color="white",edgecolor='green')
plt.title("Histogfram with KDE")
ax.set_facecolor('black')
plt.show()
```

```python
matrix=np.random.rand(5,5)
sns.heatmap(matrix,annot=True,cmap="prism")
plt.title("Heatmap")
plt.show()
```

```python
sns.set_style("whitegrid")
df=sns.load_dataset("iris")
sns.pairplot(df,hue="species", markers=["o","s","D"])
plt.show()
```

```python
import numpy as im
import matplotlib.pyplot as plt

board_size = 7  # Number of rows and columns
square_size = 500 # Size of each square in pixels

chessboard = im.zeros((board_size * square_size, board_size * square_size, 3), dtype=im.uint8)

# Fill the chessboard array with black and white squares
for row in range(board_size):
    for col in range(board_size):
        color = 255 * ((row + col) % 2)  # Alternating black and white squares
        chessboard[row * square_size: (row + 1) * square_size, col * square_size: (col + 1) * square_size, :] = color

plt.imshow(chessboard)
plt.axis('off')
plt.title("Chessboard")
plt.show()
```
